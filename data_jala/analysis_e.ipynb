```python
import pandas as pd

# Import data dari file cycles.csv
data_cycles = pd.read_csv("/workspaces/Ds_jala/Data/cycles.csv")

# Import data dari file farms.csv
data_farms = pd.read_csv("/workspaces/Ds_jala/Data/farms.csv")

# Import data dari file fasting.csv
data_fasting = pd.read_csv("/workspaces/Ds_jala/Data/fasting.csv")

# Import data dari file fastings.csv
data_fastings = pd.read_csv("/workspaces/Ds_jala/Data/fastings.csv")

# Import data dari file feed_tray.csv
data_feed_tray = pd.read_csv("/workspaces/Ds_jala/Data/feed_tray.csv")

# Import data dari file feeds.csv
data_feeds = pd.read_csv("/workspaces/Ds_jala/Data/feeds.csv")

# Import data dari file harvests.csv
data_harvests = pd.read_csv("/workspaces/Ds_jala/Data/harvests.csv")

# Import data dari file measurements.csv
data_measurements = pd.read_csv(r"/workspaces/Ds_jala/Data/measurements.csv")

# Import data dari file mortalities.csv
data_mortalities = pd.read_csv("/workspaces/Ds_jala/Data/mortalities.csv")

# Import data dari file ponds.csv
data_ponds = pd.read_csv("/workspaces/Ds_jala/Data/ponds.csv")

# Import data dari file samplings.csv
data_samplings = pd.read_csv("/workspaces/Ds_jala/Data/samplings.csv")
```

```python
# Now drop duplicates (consider inplace=True for efficiency)
for df, name in zip(
    [
        data_cycles,
        data_farms,
        data_fasting,
        data_fastings,
        data_feed_tray,
        data_feeds,
        data_harvests,
        data_measurements,
        data_mortalities,
        data_ponds,
        data_samplings,
    ],
    [
        "data_cycles",
        "data_farms",
        "data_fasting",
        "data_fastings",
        "data_feed_tray",
        "data_feeds",
        "data_harvests",
        "data_measurements",
        "data_mortalities",
        "data_ponds",
        "data_samplings",
    ],
):
    df.drop_duplicates(inplace=True)  # Drop duplicates and modify DataFrame in-place

# Print number of duplicates after dropping
print("\nAnalisis Duplikat (setelah dihapus):")
for df, name in zip(
    [
        data_cycles,
        data_farms,
        data_fasting,
        data_fastings,
        data_feed_tray,
        data_feeds,
        data_harvests,
        data_measurements,
        data_mortalities,
        data_ponds,
        data_samplings,
    ],
    [
        "data_cycles",
        "data_farms",
        "data_fasting",
        "data_fastings",
        "data_feed_tray",
        "data_feeds",
        "data_harvests",
        "data_measurements",
        "data_mortalities",
        "data_ponds",
        "data_samplings",
    ],
):
    print(f"- {name}: {df.duplicated().sum()}")  # Print number of duplicates remaining
```
