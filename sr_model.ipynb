```python
import pandas as pd
```

```python
data = pd.read_csv("data_abw.csv")
```

```python
data.head()
```

```python
data.corr()
```

```python
data = data[['bd_daya','sum_sample_g','average_weight']]
```

```python
data['sum_sample_g'] = data['sum_sample_g']/1000
```

```python
# Impor pustaka yang diperlukan
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Buat dataset contoh

X = data.drop(columns=['average_weight'], axis=1)  # Fitur
y = data['average_weight']  # Target dengan noise

# Bagi dataset menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Regresi Linier
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
y_pred_linear = linear_reg.predict(X_test)

# Regresi Polinomial
poly_features = PolynomialFeatures(degree=2)
X_poly_train = poly_features.fit_transform(X_train)
X_poly_test = poly_features.transform(X_test)

poly_reg = LinearRegression()
poly_reg.fit(X_poly_train, y_train)
y_pred_poly = poly_reg.predict(X_poly_test)

# Support Vector Regression (SVR)
svr_reg = SVR(kernel='linear')
svr_reg.fit(X_train, y_train)
y_pred_svr = svr_reg.predict(X_test)

# Random Forest Regression
rf_reg = RandomForestRegressor(n_estimators=100, random_state=0)
rf_reg.fit(X_train, y_train)
y_pred_rf = rf_reg.predict(X_test)

# Evaluasi performa model
print("Linear Regression:")
print("MSE:", mean_squared_error(y_test, y_pred_linear))
print("R-squared:", r2_score(y_test, y_pred_linear))

print("\nPolynomial Regression:")
print("MSE:", mean_squared_error(y_test, y_pred_poly))
print("R-squared:", r2_score(y_test, y_pred_poly))

print("\nSVR:")
print("MSE:", mean_squared_error(y_test, y_pred_svr))
print("R-squared:", r2_score(y_test, y_pred_svr))

print("\nRandom Forest Regression:")
print("MSE:", mean_squared_error(y_test, y_pred_rf))
print("R-squared:", r2_score(y_test, y_pred_rf))
```
